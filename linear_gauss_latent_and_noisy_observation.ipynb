```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed
np.random.seed(42)

# Parameters
L = 2 # Dimension of latent variable z 
D = 2 # Dimension of noisy observation y

# Prior over z ~ N(mu_z, Cov_z)
mu_z = np.array([0, 0])
Cov_z = np.array([[1, 0.5],[0.5, 1]])

# Linear transfomation
W = np.array([[1.0, 2.0],
              [-1.0, 0.5]])
b = np.array([0.5, -0.5])

# Observation noise covariance
Cov_y = np.array([[0.2, 0],
                  [0, 0.2]])

# Sample z from f(z)
num_samples = 300
z_samples = np.random.multivariate_normal(mu_z, Cov_z, num_samples)

# Sample y from f(y| z) = N(Wz + b, Cov_y)
y_samples = []
for z in z_samples:
    mean_y = W @ z + b
    y = np.random.multivariate_normal(mean_y, Cov_y)
    y_samples.append(y)
y_samples = np.array(y_samples)

```

```python
# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(z_samples[:,0], z_samples[:,1], alpha=0.5, color='blue', label='Latent variable z')
plt.scatter(y_samples[:,0], y_samples[:,1], alpha=0.5, color='red', label='Observation variable y (noisy)')
plt.title('Linear Gaussian System: Latent z and Noisy observation y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
